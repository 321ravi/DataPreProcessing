# import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# importing Dataset

```python
#dataset = pd.read_excel('Data.xlsx')
dataset = pd.read_excel('data.xlsx')
#dataset["target"] = dataset.shift(-1)["Maximum Temperature (2 m)"]  # added new column
dataset["target"] = dataset.shift(-1)["temperature_2m (°C)"]
dataset.head()

```

```python
dataset.drop(dataset.tail(1).index,inplace=True)
x = dataset.drop(["Date","target"],axis=1)
y = dataset["target"]


print(x)
```

# Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train)
```

# Training Model

```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=150, random_state=0)

# Convert y_train to a 1D array
#y_train_1d = y_train.ravel()

# Fit the model to the training data
#regressor.fit(x_train, y_train_1d)
regressor.fit(x_train, y_train)





```

```python
regressor.score(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)

# print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

```python
from datetime import datetime, timedelta

def preprocess_future_data(data): # return X
    data['Date'] = pd.to_datetime(data['Date'])
    data['month'] = data['Date'].dt.month
    data['day_of_year'] = data['Date'].dt.day_of_year

    
    X = data[['month', 'day_of_year' ]]
    return X

# Define the start and end dates
start_date = datetime(2024, 3, 5)
end_date = datetime(2025, 1, 1)

# Generate a list of dates from start_date to end_date
date_range = [start_date + timedelta(days=x) for x in range((end_date - start_date).days + 1)]

# Create a DataFrame with a date column
future_data = pd.DataFrame({'Date': date_range})
preprocess_future_data(future_data)

```

```python

# Assuming future_data['day_of_year'] is a 1D array or list
future_data_day_of_year = np.array(future_data['day_of_year']).reshape(-1, 1)

# Generate predictions
pred = regressor.predict(future_data_day_of_year)

```
